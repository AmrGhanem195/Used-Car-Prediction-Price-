##Data Cleaning

```python
data.isnull().sum()
```

```python
data= data.drop('Unnamed: 7', axis=1)
```

```python
data['Engine'] = data['Engine'].str.extract('(\d+\.?\d*)').astype(float)
median_engine = data['Engine'].median()
data['Engine'].fillna(median_engine, inplace=True)
```

```python
data['Engine'].sample(10)
```

```python
data['Power'] = data['Power'].str.extract('(\d+.?\d*)').astype(float)
median_Power = data['Power'].median()
data['Power'].fillna(median_Power, inplace=True)
```

```python
data['Power']
```

```python
mean_Seats = data['Seats'].mean()
data['Seats'].fillna(mean_Seats, inplace = True)
```

```python
mean_Price = data['Price'].mean()
data['Price'].fillna(mean_Price, inplace = True)
```

## Data Preprocessing

```python
#Features
x = data.iloc[ : , 0:-1 ]
#Lables
y = data.iloc [: , -1]
```

```python
data.head(10)
```

```python
y.head(10)
```

```python
def transform(t): ## To convert Transmission
  if t == 'Manual':
    return 1
  elif t == 'Automatic':
    return 2
  else:
    return t
```

```python
x['Transmission'] = x['Transmission'].apply (transform)
```

```python
data['Fuel_Type'].unique()
```

```python
def Fuel(f): ## To convert Fuel Type

  if f == 'CNG':
    return 1
  elif f == 'Diesel':
    return 2
  elif f == 'Petrol':
    return 3
  elif f == 'LPG':
    return 4
  elif f == 'Electric':
    return 5
  else:
    return f
```

```python
x['Fuel_Type'] = x['Fuel_Type'].apply(Fuel)
```

```python
x.sample(100)
```

```python
x['Owner_Type'].unique()
```

```python
def Owner_type(o): ## To convert Owner Type

  if o == 'First':
    return 1
  elif o == 'Second':
    return 2
  elif o == 'Third':
    return 3
  elif o == 'Fourth & Above':
    return 4

  else:
    return o
```

```python
x['Owner_Type'] = x['Owner_Type'].apply(Owner_type)
```

```python
x['Name'].value_counts()
```

```python
x.sample(10)
```

```python
x = pd.get_dummies(x)
```

```python
y.head(10)
```

## Scaling

```python
from sklearn.preprocessing import MinMaxScaler

scaler_x = MinMaxScaler(feature_range=(0, 1))
scaler_y = MinMaxScaler(feature_range=(0, 1))

x = scaler_x.fit_transform(x)
y = scaler_y.fit_transform(y.values.reshape(-1, 1))

```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test  , y_train , y_test = train_test_split (x , y ,test_size = 0.20)
```

```python
x_train
```

##AI

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
model1 = LinearRegression()
model1.fit(x_train , y_train)
```

```python
model1.score(x_test , y_test)*100
```
